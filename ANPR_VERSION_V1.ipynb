```python


from PIL import Image
from langchain_google_genai import ChatGoogleGenerativeAI 
from langchain_core.messages import HumanMessage



def response(image_path):
    llm = ChatGoogleGenerativeAI(model="gemini-1.5-flash-latest", google_api_key=secret_value_0)
    try:
        hmessage1 = HumanMessage(
            content=[
                {
                    "type": "text",
                    "text":'''return the number plate of the car  in this format {The number plate of the car is "Number plate".}
                    '''
                },
                {"type": "image_url", "image_url": image_path},
            ]
        )
        message1 = llm.invoke([hmessage1])
        return(message1.content)

    except:
        print("na")
```

```python
response('/kaggle/input/car-plate-detection/images/Cars153.png')
```

```python
import os
path__=[]
for i in os.listdir('/kaggle/input/car-plate-detection/images'):
    path__.append(os.path.join('/kaggle/input/car-plate-detection/images',i))
```

```python
%%time
car=[]
number=[]
for i in path__[:20]:
    
    a=response(i)
    if(a!=None):
        number.append(a)
    else:
        number.append(np.nan)
    car.append(i)
```

```python
dicto={"Path":car,
      "NUMBER":number}
```

```python
df=pd.DataFrame(dicto,columns=['Path','NUMBER'])
df.index=df.index+1
```
